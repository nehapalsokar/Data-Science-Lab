```python
#8

def gradient_ascent(x0,A,R):
    m0 = []
    l1 = []
    cnt = 0

    while True:
        m0.append(x0[0,0])
        l1.append(x0[1,0])
        x1 = x0 + (R*grad1(x0,A))
        m0.append(x1[0,0])
        l1.append(x1[1,0])
        if abs(f1(x1,A)-f1(x0,A)) < (1e-6) or cnt == 1000:
            break
        else:
            x0 = x1
            cnt = cnt + 1
  #  print (m0)
    print(cnt)
    return m0,l1;
# print(x)
# print(A)
x0_list=list()   
print('for learning rate 0.01 the number of iterations is')
gradient_ascent(x,A,0.01)
print('for learning rate 0.1 the number of iterations is')
gradient_ascent(x,A,0.1)
print('for learning rate 0.5 the number of iterations is')
gradient_ascent(x,A,0.0)


n = 30
ex1 = np.linspace(-5, 5, num=n)
ex2 = np.linspace(-5, 5, num=n)
grid1, grid2 = np.meshgrid(ex1, ex2)

## fill the grid via looping.  You may prefer to use ufuncs instead.
z = np.empty_like(grid1)
for i in range(z.shape[0]):
    for j in range(z.shape[1]):
        x = np.array([grid1[i,j], grid2[i,j]])
        z[i,j] = f1(x,A)[0][0]

## Make the plot
plt.figure(figsize=(9,9))
p = plt.contour(grid1, grid2, z,
                levels = [-50,-20, -10, -5, -2, -1, 0, 1, 2, 5, 10, 20,50],
                colors = 'black')
plt.xlim(-4,6)
plt.ylim(-3,1)
plt.clabel(p, inline=1, fontsize=10)
x0 = np.asarray(x0_list)


x0_1 = np.matrix("2;-3")
A_1 = np.matrix("1,2;2,8")
R=0.1
m0,l1 = gradient_ascent(x0_1,A_1,R)


plt.scatter(m0,l1,c='red',s=30)
plt.plot(m0,l1,c='red')

#As we can see, as the points x0 and x1 approach the maxima, they come closer and closer.
#At the maximum value, i.e. x=0, they converge.
```

```python
#Pareto:

def Pareto(N):
    R=1000
    shape, mode = 10.,2.
    values =(np.random.pareto(shape,size=(N,)) +1)* mode
    
    means = ((np.random.normal(shape, size=(N,R))+1)*mode).mean(axis=1)
    
    #calculate the mean of means
    print('Mean of means =',means.mean())
    #calculate sd
    print('SD =  ',values.std())
    #CI
    x1 = values.mean()-(1.96*values.std())
    x2 = values.mean()+(1.96*values.std())
    print("CI = " + str(x1) + " : " +str(x2))
    
Pareto(10)
Pareto(100)
Pareto(1000)
```
